```python
#If there is any emojis, then it will be convert to N/A
import re

emoji_pattern = re.compile(
    "["
    "\U0001F600-\U0001F64F"  # Emoticons
    "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
    "\U0001F680-\U0001F6FF"  # Transport & Map
    "\U0001F1E0-\U0001F1FF"  # Flags
    "\U00002700-\U000027BF"  # Dingbats
    "\U000024C2-\U0001F251"
    "]+",
    flags=re.UNICODE
)

all_data = all_data.apply(
    lambda col: col.map(
        lambda x: emoji_pattern.sub("N/A", x) if isinstance(x, str) else x
    )
)
```
