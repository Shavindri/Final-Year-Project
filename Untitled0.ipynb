```python
# Categorise scenario responses automatically

scenario_chart = pd.DataFrame(index=[
    "Correct",
    "Incorrect / Risky",
    "Uncertain / Partial"
])

# Email Scenario
email = all_data["Scenario_Email"].value_counts()

scenario_chart["Phishing Email"] = [
    email.get("Verify the request using the bank's official website or customer service", 0),
    email.get("Click the link immediately", 0),
    email.get("Ignore it completely", 0)
]

# OTP Scenario
otp = all_data["Scenario_OTP"].value_counts()

scenario_chart["OTP Request"] = [
    otp.get("End the call and contact the bank through official channels", 0),
    otp.get("Share the OTP", 0),
    otp.get("Ignore the request but continue the conversation", 0)
        + otp.get("Ask why it is needed", 0)
]

# Login Alert Scenario
login = all_data["Scenario_Login_Alert"].value_counts()

scenario_chart["Login Alert"] = [
    login.get("Change your password immediately and contact the bank", 0),
    login.get("Ignore it", 0),
    login.get("Wait to see if it happens again", 0)
]

# Deepfake Call Scenario
deepfake = all_data["Scenario_Deepfake_Call"].value_counts()

scenario_chart["Deepfake Call"] = [
    deepfake.get("Verify the request using the bank's official contact details", 0),
    deepfake.get("Approve the transaction", 0),
    deepfake.get("End the call without checking", 0)
]

# Plot graph
scenario_chart.T.plot(
    kind="bar",
    figsize=(10,6),
    edgecolor="black"
)

plt.title("Scenario-Based Security Response Analysis")
plt.xlabel("Scenario")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)
plt.legend(title="Response Type")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(all_data["Age"], all_data["Awareness_Level"])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", round(chi2,3))
print("Degrees of Freedom:", dof)
print("P-value:", round(p,3))
```

```python
from scipy.stats import pearsonr

pairs = [
    ("Awareness_Score", "Opinion_Score"),
    ("Awareness_Score", "Behaviour_Score"),
    ("Awareness_Score", "Biometric_Score"),
    ("Opinion_Score", "Behaviour_Score"),
    ("Opinion_Score", "Biometric_Score"),
    ("Behaviour_Score", "Biometric_Score")
]

results = []

for x, y in pairs:

    r, p = pearsonr(all_data[x], all_data[y])

    if abs(r) < 0.20:
        strength = "Very Weak"
    elif abs(r) < 0.40:
        strength = "Weak"
    elif abs(r) < 0.60:
        strength = "Moderate"
    elif abs(r) < 0.80:
        strength = "Strong"
    else:
        strength = "Very Strong"

    significance = "Significant" if p < 0.05 else "Not Significant"

    results.append({
        "Variable 1": x,
        "Variable 2": y,
        "Correlation (r)": round(r,3),
        "P-value": round(p,3),
        "Strength": strength,
        "Decision": significance
    })

results_df = pd.DataFrame(results)

print(results_df)

results_df.to_csv("Correlation_Results.csv", index=False)
```

```python
from scipy.stats import shapiro

variables = [
    "Awareness_Score",
    "Opinion_Score",
    "Behaviour_Score",
    "Biometric_Score"
]

for col in variables:
    stat, p = shapiro(all_data[col])

    print(f"\n{col}")
    print("Statistic =",round(stat,3))
    print("P-value =",round(p,3))
```

```python
encoder = LabelEncoder()

categorical = [
    "Age",
    "Gender",
    "Education",
    "Banking_Use",
    "Awareness_Level"
]

for col in categorical:
    all_data[col]=encoder.fit_transform(all_data[col])
```

```python
prediction=model.predict(X_test)
```
