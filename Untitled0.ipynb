```python
import pandas as pd
import matplotlib.pyplot as plt

# Categorise scenario responses
scenario_chart = pd.DataFrame(index=[
    "Correct",
    "Incorrect / Risky",
    "Uncertain / Partial"
])

# Email Scenario
email = all_data["Scenario_Email"].value_counts()

scenario_chart["Phishing Email"] = [
    email.get(
        "Verify the request using the bank's official website or customer service",
        0
    ),
    email.get("Click the link immediately", 0),
    email.get("Ignore it completely", 0)
]

# OTP Scenario
otp = all_data["Scenario_OTP"].value_counts()

scenario_chart["OTP Request"] = [
    otp.get(
        "End the call and contact the bank through official channels",
        0
    ),
    otp.get("Share the OTP", 0),
    otp.get("Ignore the request but continue the conversation", 0)
    + otp.get("Ask why it is needed", 0)
]

# Login Alert Scenario
login = all_data["Scenario_Login_Alert"].value_counts()

scenario_chart["Login Alert"] = [
    login.get(
        "Change your password immediately and contact the bank",
        0
    ),
    login.get("Ignore it", 0),
    login.get("Wait to see if it happens again", 0)
]

# Deepfake Scenario
deepfake = all_data["Scenario_Deepfake_Call"].value_counts()

scenario_chart["Deepfake Call"] = [
    deepfake.get(
        "Verify the request using the bank's official contact details",
        0
    ),
    deepfake.get("Approve the transaction", 0),
    deepfake.get("End the call without checking", 0)
]

# Convert table so scenarios appear on the x-axis
scenario_plot = scenario_chart.T

# Create chart
ax = scenario_plot.plot(
    kind="bar",
    figsize=(11, 6),
    width=0.75,
    edgecolor="black",
    linewidth=0.8
)

# Add number and percentage labels
total_respondents = len(all_data)

for container in ax.containers:
    labels = []

    for bar in container:
        value = bar.get_height()

        if value > 0:
            percentage = (value / total_respondents) * 100
            labels.append(f"{int(value)}\n({percentage:.1f}%)")
        else:
            labels.append("")

    ax.bar_label(
        container,
        labels=labels,
        padding=4,
        fontsize=9,
        fontweight="bold"
    )

# Chart formatting
plt.title(
    "Scenario-Based Online Banking Security Responses",
    fontsize=15,
    fontweight="bold",
    pad=40
)

plt.xlabel(
    "Security Scenario",
    fontsize=11,
    fontweight="bold"
)

plt.ylabel(
    "Number of Respondents",
    fontsize=11,
    fontweight="bold"
)

plt.xticks(
    rotation=0,
    fontsize=10
)

plt.yticks(fontsize=10)

plt.legend(
    title="Response Classification",
    loc="upper center",
    bbox_to_anchor=(0.5, 1.16),
    ncol=3,
    frameon=False
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

plt.ylim(
    0,
    scenario_plot.values.max() + 7
)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(all_data["Age"], all_data["Awareness_Level"])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", round(chi2,3))
print("Degrees of Freedom:", dof)
print("P-value:", round(p,3))
```

```python
from scipy.stats import pearsonr

pairs = [
    ("Awareness_Score", "Opinion_Score"),
    ("Awareness_Score", "Behaviour_Score"),
    ("Awareness_Score", "Biometric_Score"),
    ("Opinion_Score", "Behaviour_Score"),
    ("Opinion_Score", "Biometric_Score"),
    ("Behaviour_Score", "Biometric_Score")
]

results = []

for x, y in pairs:

    r, p = pearsonr(all_data[x], all_data[y])

    if abs(r) < 0.20:
        strength = "Very Weak"
    elif abs(r) < 0.40:
        strength = "Weak"
    elif abs(r) < 0.60:
        strength = "Moderate"
    elif abs(r) < 0.80:
        strength = "Strong"
    else:
        strength = "Very Strong"

    significance = "Significant" if p < 0.05 else "Not Significant"

    results.append({
        "Variable 1": x,
        "Variable 2": y,
        "Correlation (r)": round(r,3),
        "P-value": round(p,3),
        "Strength": strength,
        "Decision": significance
    })

results_df = pd.DataFrame(results)

print(results_df)

results_df.to_csv("Correlation_Results.csv", index=False)
```

```python
from scipy.stats import shapiro

variables = [
    "Awareness_Score",
    "Opinion_Score",
    "Behaviour_Score",
    "Biometric_Score"
]

for col in variables:
    stat, p = shapiro(all_data[col])

    print(f"\n{col}")
    print("Statistic =",round(stat,3))
    print("P-value =",round(p,3))
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd

st.set_page_config(
    page_title="Cybersecurity in Online Banking",
    layout="wide"
)

df = pd.read_csv("Online_Banking_Cleaned_Dataset.csv")

st.title("Cybersecurity Awareness in Online Banking")

page = st.sidebar.selectbox(
    "Menu",
    [
        "Home",
        "Dataset Explorer",
        "Data Analysis",
        "Security Assessment",
        "Tips & Resources"
    ]
)

# Home page
if page == "Home":

    st.header("Welcome")

    st.write("""
    Welcome to the Online Banking Cybersecurity Awareness Toolkit. Developed
    from the findings of this research survey, the toolkit helps users explore
    cybersecurity risks, assess their security awareness, and learn recommended
    online banking practices. \\\
    It is designed for educational purposes only and
    is not a banking authentication or financial transaction system.
    """)

# Dataset page
elif page == "Dataset Explorer":

    st.header("Dataset Explorer")

    st.write("Preview of the cleaned survey dataset:")

    st.dataframe(df.head())

    st.write("Dataset size:", df.shape)

    st.write("Number of respondents:", df.shape[0])

    st.write("Number of variables:", df.shape[1])

# Data analysis page
elif page == "Data Analysis":

    st.header("Survey Findings and Data Analysis")

    chart_option = st.selectbox(
        "Select an analysis:",
        [
            "Age Distribution",
            "Awareness Levels",
            "Phishing Experiences",
            "Security Behaviours",
            "Opinions on Security Measures",
            "Main Biometric Concerns"
        ]
    )

    if chart_option == "Age Distribution":

        st.subheader("Age Distribution")
        st.bar_chart(df["Age"].value_counts())

    elif chart_option == "Awareness Levels":

        st.subheader("Average Cybersecurity Awareness Scores")

        awareness_columns = [
            "Aware_Phishing",
            "Aware_OTP",
            "Aware_MFA",
            "Aware_Website",
            "Aware_PublicWiFi",
            "Aware_AI",
            "Aware_Response"
        ]

        awareness_means = df[awareness_columns].mean().sort_values()
        st.bar_chart(awareness_means)

        st.info(
            "Lower mean scores indicate stronger awareness because "
            "1 represents Strongly Agree and 5 represents Strongly Disagree."
        )

    elif chart_option == "Phishing Experiences":

        st.subheader("Experience with Phishing Emails")
        st.bar_chart(df["Exp_Phishing_Email"].value_counts())

    elif chart_option == "Security Behaviours":

        st.subheader("Average Security Behaviour Scores")

        behaviour_columns = [
            "Behaviour_UniquePassword",
            "Behaviour_UseMFA",
            "Behaviour_UpdateApp",
            "Behaviour_VerifyWebsite"
        ]

        behaviour_means = df[behaviour_columns].mean().sort_values()
        st.bar_chart(behaviour_means)

    elif chart_option == "Opinions on Security Measures":

        st.subheader("Opinions on MFA and Biometric Authentication")

        opinion_means = df[
            [
                "Opinion_MFA",
                "Opinion_Biometrics"
            ]
        ].mean()

        st.bar_chart(opinion_means)

    elif chart_option == "Main Biometric Concerns":

        st.subheader("Main Concerns About Biometric Authentication")
        st.bar_chart(df["Biggest_Concern"].value_counts())

# Security assessment page
elif page == "Security Assessment":

    st.header("Cybersecurity Awareness Assessment")

    st.write(
        "Answer the following questions to receive a basic awareness result."
    )

    phishing = st.selectbox(
        "Can you recognise phishing emails?",
        ["Select an answer", "Yes", "Sometimes", "No"]
    )

    otp = st.selectbox(
        "Would you share your OTP with someone claiming to be from your bank?",
        ["Select an answer", "No", "Not Sure", "Yes"]
    )

    score = 0

    if phishing == "Yes":
        score += 1
    elif phishing == "Sometimes":
        score += 0.5

    if otp == "No":
        score += 1
    elif otp == "Not Sure":
        score += 0.5

    if st.button("Check My Awareness"):

        if phishing == "Select an answer" or otp == "Select an answer":

            st.warning("Please answer all questions.")

        elif score == 2:

            st.success("High Awareness")

            st.write(
                "You demonstrate a good understanding of phishing and OTP safety."
            )

        elif score >= 1:

            st.warning("Moderate Awareness")

            st.write(
                "You understand some important security practices, but further "
                "learning is recommended."
            )

        else:

            st.error("Low Awareness")

            st.write(
                "You may be vulnerable to common online banking threats. "
                "Please review the Tips & Resources page."
            )

    st.caption(
        "This assessment is for educational purposes only and is not a formal "
        "security or banking-authentication test."
    )

# Tips page
elif page == "Tips & Resources":

    st.header("Cybersecurity Tips & Resources")

    st.markdown("""
    ### Protect your online banking account

    ✅ Never share your password, PIN or OTP.

    ✅ Enable Multi-Factor Authentication whenever available.

    ✅ Check website addresses before entering banking information.

    ✅ Avoid accessing online banking through public Wi-Fi.

    ✅ Keep your banking applications and mobile device updated.

    ✅ Do not click links in unexpected banking emails or messages.

    ✅ Contact your bank through its official website or telephone number.

    ✅ Report suspicious transactions immediately.
    """)
```
