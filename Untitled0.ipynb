```python
awareness = [
    "Aware_Phishing",
    "Aware_OTP",
    "Aware_MFA",
    "Aware_Website",
    "Aware_PublicWiFi",
    "Aware_AI",
    "Aware_Response"
]

for col in awareness:
    summary = pd.DataFrame({
        "Frequency": all_data[col].value_counts().sort_index(),
        "Percentage": round(all_data[col].value_counts(normalize=True).sort_index() * 100, 2)
    })

    print("\n", col)
    print(summary)
```

```python
#Calculate the Mean and Standard Deviation
awareness_summary = pd.DataFrame({
    "Mean": round(all_data[awareness].mean(), 2),
    "Standard Deviation": round(all_data[awareness].std(), 2)
})

awareness_summary
```

```python
all_data["Awareness_Score"] = round(all_data[awareness].mean(axis=1), 2)

all_data["Awareness_Score"].describe()
#Calculate an Overall Awareness Score
```

```python
#Categorise Respondents by Awareness Level
def awareness_level(score):
    if score <= 2:
        return "High Awareness"
    elif score <= 3:
        return "Moderate Awareness"
    else:
        return "Low Awareness"

all_data["Awareness_Level"] = all_data["Awareness_Score"].apply(awareness_level)

all_data["Awareness_Level"].value_counts()
```

```python
import matplotlib.pyplot as plt

all_data["Awareness_Level"].value_counts().plot(
    kind="bar",
    edgecolor="black"
)

plt.title("Overall Cybersecurity Awareness Level")
plt.xlabel("Awareness Level")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=0)

plt.show()
```
